```python
import glob
import numpy as np
from numpy import genfromtxt
import os
from tqdm import tqdm
import sys
sys.path.append('../')
from utils import project_root

def LZ76(ss):
    """
    Calculate Lempel-Ziv's algorithmic complexity using the LZ76 algorithm
    and the sliding-window implementation.

    Reference:

    F. Kaspar, H. G. Schuster, "Easily-calculable measure for the
    complexity of spatiotemporal patterns", Physical Review A, Volume 36,
    Number 2 (1987).

    Input:
      ss -- array of integers

    Output:
      c  -- integer
    """

    ss = ss.flatten().tolist()
    i, k, l = 0, 1, 1
    c, k_max = 1, 1
    n = len(ss)
    while True:
        if ss[i + k - 1] == ss[l + k - 1]:
            k = k + 1
            if l + k > n:
                c = c + 1
                break
        else:
            if k > k_max:
               k_max = k
            i = i + 1
            if i == l:
                c = c + 1
                l = l + k_max
                if l + 1 > n:
                    break
                else:
                    i = 0
                    k = 1
                    k_max = 1
            else:
                k = 1
    return c

root = project_root()

for f in tqdm(glob.iglob(os.path.join(root, 'time_series/*/*.txt'), recursive=True)):
    dim = 116 if 'aal' in f else 100
        
    ts = genfromtxt(f, delimiter=' ').reshape(dim, -1)

    lzs = np.zeros((dim))
    
    for i, X in enumerate(ts):
        print(X.shape)
        Y = X > np.median(X)
        print(X)
        print(Y.shape)
        print(np.median(X))
        lzs[i] = LZ76(Y)*np.log2(len(Y))/len(Y)
        print(lzs[i])

    # save the array to a CSV file
    
    p1 = f.split("/")[-2]
    p2 = f.split('/')[-1].split(".")[0]
#     np.savetxt(os.path.join(root, f'lempel_ziv/{p1}/{p2}.csv'), lzs, delimiter=',')

# # root_dir needs a trailing slash (i.e. /root/dir/)
# for f in tqdm(glob.iglob(os.path.join(root, 'time_series/*/*.csv'), recursive=True)):
#     dim = 116 if 'aal' in f else 100
        
#     ts = genfromtxt(f, delimiter=',').reshape(dim, -1)

#     lzs = np.zeros((dim))
#     for i, X in enumerate(ts):
#         Y = X > np.median(X)
#         lzs[i] = LZ76(Y)*np.log2(len(Y))/len(Y)

#     # save the array to a CSV file
    
#     p1 = f.split("/")[-2]
#     p2 = f.split('/')[-1].split(".")[0]
#     np.savetxt(os.path.join(root, f'lempel_ziv/{p1}/{p2}.csv'), lzs, delimiter=',')
```
